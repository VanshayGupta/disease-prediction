```python
heart_list = ["heart", "ventricular", "carotid",
              "cardiac", "angina", "myocardial",
              "coronary", "cardio", "aortic"]
heart_freq=df.loc[df['msdrgcode'].str
                  .contains('|'.join(heart_list),
                            case=False)]
heart_freq
heart_unique = pd.DataFrame(df["msdrgcode"].unique())
heart_freq['freq_occurance'] = heart_freq.groupby('msdrgcode')
                               .cumcount() + 1
heart_freq= heart_freq.reset_index()
freq = heart_freq.groupby('msdrgcode').size()
for ind, i in enumerate(heart_freq['msdrgcode']):  
  if(i.find(" W")!= -1):
    heart_freq['msdrgcode'][ind] = 
    heart_freq['msdrgcode'][ind][: i.find(" W")]
freq = heart_freq.groupby('msdrgcode').size()
freq
```
